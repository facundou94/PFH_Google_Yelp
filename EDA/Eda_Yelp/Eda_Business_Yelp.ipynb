Un archivo .pkl es un archivo binario utilizado en Python para almacenar estructuras de datos serializadas utilizando el módulo pickle. El nombre "pickle" proviene del inglés y hace referencia a preservar o conservar algo para su uso posterior.

```python
# Calcular valores nulos y porcentajes
valores_nulos = business_california.isnull().sum()
porcentajes = (valores_nulos / business_california.size) * 100  # Calcula el porcentaje respecto al total de elementos del DataFrame

# Eliminar columnas sin valores nulos (opcional)
porcentajes = porcentajes[porcentajes != 0]

# Ordenar por porcentaje descendente para un mejor gráfico
porcentajes = porcentajes.sort_values(ascending=False)

# Mostrar porcentajes calculados
print("Porcentajes de valores nulos por columna:")
print(valores_nulos)
```

```python
# Convertir 'review_count' a tipo int si aún no lo está
business_california['review_count'] = business_california['review_count'].astype(int)

# Top 10 negocios con más reseñas
top_n_negocios = 10
top_negocios = business_california.nlargest(top_n_negocios, 'review_count')
print(f"Top {top_n_negocios} negocios con más reseñas:")
print(top_negocios[['name', 'review_count']])

# Gráfico de barras de los 10 negocios con más reseñas
plt.figure(figsize=(10, 6))
plt.barh(top_negocios['name'], top_negocios['review_count'], color='skyblue')
plt.xlabel('Review Count')
plt.ylabel('Business')
plt.title('Top 10 Businesses by Review Count')
plt.gca().invert_yaxis()  # Invertir eje y para que el negocio con más reseñas esté arriba
plt.show()
```

```python
# Conteo de las principales categorías de negocios
top_categorias = business_california['categories'].value_counts().nlargest(10)
print("Top 10 categorías de negocios:")
print(top_categorias)

# Gráfico de barras de las 10 principales categorías de negocios
plt.figure(figsize=(10, 6))
top_categorias.plot(kind='bar', color='lightgreen')
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Top 10 Business Categories')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python



df = pd.DataFrame(business_california)

keywords = {
    'food': ['food','Restaurants'],
    'shopping': ['shopping','Accessories'],
    'entretenimiento': ['entertainment'],
    'hospital': ['hospitals']
}

def assign_group(categoría):
    if pd.isna(categoría):
        return 'other'  # Si el valor es None o NaN, asigna 'other'
    for group, words in keywords.items():
        if any(word in categoría.lower() for word in words):
            return group
    return 'other'  # Si no coincide con ninguna palabra clave, asigna 'other'


df['grupo'] = df['categories'].apply(assign_group)

group_counts = df['grupo'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(group_counts, labels=group_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Categorías')
plt.show()
```
